```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

# Load the uploaded Excel file
for Texasgovcomp in uploaded.keys():
    df = pd.read_excel(Texasgovcomp) # Changed to pd.read_excel

df.head()
```

```python
df.columns = df.columns.str.strip()

```

```python
df['FULL_NAME'] = df['FIRST NAME'].str.strip() + \
                  ' ' + df['MI'].fillna('').str.strip() + \
                  ' ' + df['LAST NAME'].str.strip()

# Clean double spaces (for missing middle initial)
df['FULL_NAME'] = df['FULL_NAME'].str.replace('  ', ' ')
```

```python
df = df.drop(columns=['FIRST NAME', 'MI', 'LAST NAME'])
```

```python
# Get current columns
cols = df.columns.tolist()

# Remove FULL_NAME from wherever it is
cols.remove('FULL_NAME')

# Insert FULL_NAME at position index 2 (3rd column)
cols.insert(2, 'FULL_NAME')

# Reassign ordered columns
df = df[cols]
```

```python
df.head()       # first 5 rows
df.tail()       # last 5 rows
df.sample(10)   # random sample
```

```python
df['HIRE DATE'] = pd.to_datetime(df['HIRE DATE'], errors='coerce')
#fix dates
```

```python
df['TENURE_YEARS'] = (pd.Timestamp.today() - df['HIRE DATE']) / pd.Timedelta(days=365)
#tenured
```

```python
for col in ['ETHNICITY', 'GENDER', 'EMPLOYEE TYPE', 'CLASS TITLE', 'AGENCY NAME']:
    df[col] = df[col].astype(str).str.strip().str.upper()
# normalized text
```

```python
df['EMP_TYPE_CODE'] = df['EMPLOYEE TYPE'].str.split('-').str[0].str.strip()
#extract employment type
```

```python
num_cols = ['RATE', 'HRSWKD', 'MONTHLY', 'ANNUAL', 'summed_annual_salary']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
#numerized salary
```

```python
#stndardized ethnicity
df['ETHNICITY'] = df['ETHNICITY'].replace({
    'HISPANIC/LATINO': 'HISPANIC',
    'LATINO': 'HISPANIC',
    'WHITE (NON-HISPANIC)': 'WHITE'
})
```

```python
df = df.drop_duplicates(subset=['FULL_NAME', 'AGENCY NAME'])
#remove duplicates
```

```python
#rearrange
df = df[['AGENCY', 'AGENCY NAME', 'FULL_NAME', 'CLASS TITLE', 'EMPLOYEE TYPE',
         'ETHNICITY', 'GENDER', 'HIRE DATE', 'TENURE_YEARS',
         'RATE', 'HRSWKD', 'MONTHLY', 'ANNUAL', 'summed_annual_salary']]
```
