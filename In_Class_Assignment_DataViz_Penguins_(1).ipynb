## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
# Scatter plot of bill length vs bill depth, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='Set2',
    s=80
)
plt.title('Bill Length vs Bill Depth of Penguins by Species', fontsize=14)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()

```

**Your interpretation (2–3 sentences):**

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
# Histogram with 10 bins
plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='bill_length_mm', bins=10, color='skyblue', edgecolor='black')
plt.title('Bill Length Distribution (10 bins)', fontsize=14)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()


```

```python


# Histogram with 30 bins
plt.figure(figsize=(8, 5))
sns.histplot(data=penguins, x='bill_length_mm', bins=30, color='orange', edgecolor='black')
plt.title('Bill Length Distribution (30 bins)', fontsize=14)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

We can notice bigger variations with the histogram of 30 bin. The first one is more uniform as well.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

# Boxplot of body mass grouped by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='Set3')
plt.title('Body Mass of Penguins by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```
